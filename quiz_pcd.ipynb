```python
import numpy as np
from PIL import Image
import io
from google.colab import files

uploaded = files.upload()

for filename in uploaded.keys():
    print(f'File {filename} berhasil diunggah.')

# Membaca gambar yang diunggah
uploaded_image = list(uploaded.keys())[0]  # Mengambil nama file gambar yang diunggah
image_data = io.BytesIO(uploaded[uploaded_image])
image = Image.open(image_data)

# Konversi gambar ke array NumPy
image_array = np.array(image)

# Menghitung dimensi gambar
height, width, _ = image_array.shape

# Konversi gambar ke grayscale
grayscale_image_array = np.dot(image_array[..., :3], [0.2989, 0.5870, 0.1140]).astype(np.uint8)

# Mengubah array NumPy kembali menjadi gambar PIL
grayscale_image = Image.fromarray(grayscale_image_array)

# Menyimpan gambar grayscale (opsional)
grayscale_image.save('grayscale_image.jpg')

# Menampilkan gambar asli
display(image)

# Menampilkan dimensi gambar
print(f"Lebar gambar: {width} piksel")
print(f"Tinggi gambar: {height} piksel")
```
